```python
import pandas as pd
```

```python
filepath="DataFiles/car_sales.csv"
df=pd.read_csv(filepath)
```

```python
df.isnull()
#to find rows with missing values it will bring true
```

```python
df.isnull().sum()
#returns total number of missing values per column
```

```python
df[df.isnull().any(axis=1)]
#used to find exact value missing 
```

```python
df.fillna(value={'Manufacturer':'Unknown'},inplace=True)
df.head(10)
#replaces the blank space with unknown
```

```python
average=df['Sales'].mean ()
df.fillna(value={'Sales':average},inplace=True)
df.head(10)
```

```python
average=df['Revenue'].mean ()
df.fillna(value={'Revenue':'average'},inplace=True)
df.head(10)
```

```python
df.dropna(inplace=True) #drop columns that you dont need 
df.head(10)
```

```python
df[df.duplicated()]#drop duplicates 
```

```python
df.drop_duplicates(inplace=True)#drop duplicates 
df.tail()

```

```python
#identifying outliers
q1=df["Sales"].quantile(0.25)
q3=df["Sales"].quantile(0.75)
iqr=q3-q1
print(q1,q3,iqr)
```

```python
lower_bond=q1-1.5*iqr
lower_bond
```

```python
upper_bond=q3+1.5*iqr
upper_bond 
```

```python
required=df[(df["Sales"] >=lower_bond) & (df["Sales"]<=upper_bond)]
print("All Sales/n",df["Sales"].values)
print("Sales after removing outliers/n", required["Sales"].values)
```
