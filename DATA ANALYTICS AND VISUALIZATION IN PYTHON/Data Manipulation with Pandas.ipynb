3.5 Grouping and Aggregation
Grouping Data To group data by the entity and calculate aggregate statistics:

Explanation of Aggregation Functions
sum: Total sum of values within each group.
mean: Average value of each group.
count: Number of non-null values in each group.
min: Minimum value in each group.
max: Maximum value in each group.
std: Standard deviation, indicating variability.
var: Variance, the square of the standard deviation.
median: The middle value, not affected by outliers.


```python

# Group by 'entity'
grouped = df.groupby("Entity")
print(grouped)
```

```python
# Example: Calculate mean LifeExpectancy for each entity
mean_life_expectancy = df.groupby("Entity")["LifeExpectancy"].mean()
mean_life_expectancy
```

```python
# Example: Calculate sum of LifeExpectancy for each entity
sum_life_expectancy = df.groupby("Entity")["LifeExpectancy"].sum()
sum_life_expectancy
```

```python
# Multiple aggregation functions
agg_results = df.groupby('Entity').agg({
    'LifeExpectancy': ['sum', 'mean', 'count', 'min', 'max', 'std', 'var', 'median']
})
agg_results
```
