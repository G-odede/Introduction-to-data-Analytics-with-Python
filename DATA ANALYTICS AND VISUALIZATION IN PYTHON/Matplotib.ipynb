```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data.xlsx')
df.head()
```

```python
# Aggregated data by car model
Sales_data = df.groupby('Category')['Sales'].sum()
Sales_data
```

```python
#create the pir chart
plt.pie(Sales_data,labels=Sales_data.index,autopct='%3.3f%%')
plt.title('proportional of sales by category')
```

```python
plt.hist(df['Sales'],bins=7,color='darkblue',edgecolor='black')
plt.title('Proportional of sales by category')
plt.xlabel('sales')
plt.ylabel('frequncy')
plt.show()
```

```python
#plotting data
plt.fill_between(df['Date'],df['Sales'],color='lightgreen',alpha=0.3)
plt.plot(df['Date'],df['Sales'],color='green')
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
